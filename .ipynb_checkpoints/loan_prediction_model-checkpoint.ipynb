```python
df["previous_loan_defaults_on_file"]=df["previous_loan_defaults_on_file"].replace({
    "Yes": 1,
    "No":0
})
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:13],df.iloc[:,-1],test_size=0.2)
```

```python
print(f"Train size: {x_train.shape}, Test size: {x_test.shape}")
```

```python
x_train
```

```python
duplicates = pd.merge(x_train, x_test, how='inner')
print(f"Duplicate rows between train/test: {len(duplicates)}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
    ("tnf1",OneHotEncoder(drop="first",sparse_output=False,handle_unknown="ignore"),["person_gender","person_home_ownership","loan_intent"]),
    ("tnf2",OrdinalEncoder(categories=[["High School","Associate","Bachelor","Master","Doctorate"]]),["person_education"]),
    ("tnf3",StandardScaler(),["person_income","loan_amnt","credit_score"])
],remainder="passthrough")
```

```python
x_train=transformer.fit_transform(x_train)
x_test=transformer.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000)
```

```python
lr.fit(x_train,y_train)
```

```python
prediction=lr.predict(x_test)
prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,prediction)*100
```

```python
print(f"Accuracy : {accuracy} %")
```

```python
train_acc = lr.score(x_train, y_train)  # This will be high — don't use this
test_acc  = lr.score(x_test, y_test)    # This is the real accuracy

print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy:  {test_acc:.4f}")
```

```python
# Check how many features the model actually has
print(f"Model features: {lr.coef_[0].shape[0]}")

# If you used OneHotEncoder or get_dummies, get names like this
print(f"X_train columns: {x_train.shape[1]}")
```

```python
feature_names = [
    'person_age',
    'person_gender_male',
    'person_education',
    'person_income',
    'person_emp_exp',
    'person_home_ownership_MORTGAGE',
    'person_home_ownership_OTHER',
    'person_home_ownership_OWN',
    'loan_amnt',
    'loan_intent_DEBTCONSOLIDATION',
    'loan_intent_EDUCATION',
    'loan_intent_HOMEIMPROVEMENT',
    'loan_intent_MEDICAL',
    'loan_intent_VENTURE',
    'loan_int_rate',
    'loan_percent_income',
    'cb_person_cred_hist_length',
    'credit_score',
    "previous_loan_defaults_on_file"
]

importances = pd.Series(
    np.abs(lr.coef_[0]),
    index=feature_names
).sort_values(ascending=False)

print(importances)
```

```python
import pickle

# Save
with open("transformer.pkl", "wb") as f:
    pickle.dump(transformer, f)

with open("loan_prediction_model.pkl", "wb") as f:
    pickle.dump(lr, f)

print("Saved.")

# Immediately verify they load correctly, right here, before moving them anywhere
with open("transformer.pkl", "rb") as f:
    test_transformer = pickle.load(f)

with open("loan_prediction_model.pkl", "rb") as f:
    test_model = pickle.load(f)

print("Both files loaded successfully:", type(test_transformer), type(test_model))
```
